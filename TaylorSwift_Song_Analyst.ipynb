```python
import pandas as pd
import requests
from io import StringIO

# Corrected URL of the raw CSV file on GitHub
url = "https://raw.githubusercontent.com/Jeniejean/ANALYTICS-AND-TOOLS-AND-PROGRAMMING/main/taylor_swift_songs.csv"

# Fetch the content of the CSV file
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Read the CSV data into a DataFrame
    df = pd.read_csv(StringIO(response.text))
    # Display the DataFrame
    print(df.head())
else:
    print("Failed to fetch data from URL:", url)
```
