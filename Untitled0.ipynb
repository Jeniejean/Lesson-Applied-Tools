```python
import requests

# กำหนด URL ของ API
api_url = "https://api.binance.com/api/v3/ticker?symbol=BTCUSDT"

# ทำการร้องขอข้อมูล
response = requests.get(api_url)

# ตรวจสอบว่าร้องขอสำเร็จหรือไม่
if response.status_code == 200:
    # ถ้าสำเร็จให้แปลงข้อมูล JSON เป็น Python dictionary
    data = response.json()
    # ตรวจสอบว่ามีคีย์ 'lastPrice' ใน dictionary หรือไม่
    if 'lastPrice' in data:
        # ถ้ามีให้แสดงผลราคาล่าสุด
        last_price = float(data['lastPrice'])
        print("Last Price:", last_price)
    else:
        print("ไม่พบข้อมูลราคาล่าสุด")
else:
    print("ไม่สามารถเชื่อมต่อกับ API ได้")
```
