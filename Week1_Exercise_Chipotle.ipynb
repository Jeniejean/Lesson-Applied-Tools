### Step 8. How many products cost more than $10.00?
ans 31

```python
#prices = []
if df.item_price[0] == '$':
  prices = [float(value[1:-1]) for value in df.item_price]
df.item_price = prices
df2 = df.query('item_price > 10')
print(df2.item_name.nunique()) #remove duplicate item name
df2.shape
```
