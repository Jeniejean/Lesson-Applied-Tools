### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the first dataset [cars1](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv) and [cars2](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv).  

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv", sep=",")
df2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv", sep=",")

# Display the first few rows of df1
df1.head()
```

   ### Step 3. Assign each to a to a variable called cars1 and cars2

```python
cars1 = df1
cars2 = df2
```

```python

print(cars1.head())
print(cars2.head())
```

## Step 3.1 Show cars1 information

## Step 3.2 Show cars2 information

### Step 4. Oops, it seems our first dataset has some unnamed blank columns, fix cars1

```python
cars1 = cars1.dropna(axis=1)
cars1
```

### Step 5. What is the number of observations in each dataset?

```python
print(cars1.shape)
print(cars2.shape)
```

### Step 6. Join cars1 and cars2 into a single DataFrame called cars

```python
cars = pd.concat([cars1,cars2])
cars
```
